### 三次関数にフィットしたカラーマップ

```python
data = hed2array(path, 0, 1) # 計測データの読み込み
peak_index = detect_peak_neg(data) # 64電極すべてのピーク抽出
popts, r2s = remove_fit_data(data, peak_index, 450) # 正しくピークを検出できなかったchを除外して3次関数にフィットする
draw_2d(popts, 450, 100) # 二次元カラーマップを描画
draw_3d(popts, 450, 100) # 三次元カラーマップを描画
```
